```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "My name is Kamran."
results = embeddings.embed_query(text)
results
```
