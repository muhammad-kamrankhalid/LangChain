```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
```

```python
model = ChatGroq(model="Gemma2-9b-It", groq_api_key=groq_api_key)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
messages=[
    SystemMessage(content="You are an chatbot that answers to questions with detail"),
    HumanMessage(content="What is Gen AI?")
]

result=model.invoke(messages)
```

```python
result
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)  # 'Bonjour, comment ça va ?'
```

```python
# chaining components together
chain = model | parser
chain.invoke(messages)  
```

```python
# template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant that translates following to {language}"),
        ("human", "{text}")
    ]
)
```

```python
result=prompt.invoke({"language":"Urdu","text":"Hello"})
```

```python
result.to_messages()
```

```python
result
```

```python
##Chaining together components with LCEL
chain=prompt|model|parser
chain.invoke({"language":"Urdu","text":"How are you?"})
```
